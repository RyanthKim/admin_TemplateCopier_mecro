```python
# 1.어드민 엑세스 토큰 받는곳

URL = 'https://identityaccess.api.modusign.co.kr/v1/admin/users/'
TOKEN_URL = 'https://identityaccess.api.modusign.co.kr/v1/user/oauth/token'
X_API_KEY = 'FidrahieciWU9wQrieCY4rjODW5WGIbaeFr8Li8e'
data = json.dumps(credential)
headers = {
    "x-api-key": X_API_KEY,
    "contentType": "application/json",
    "authorization": "none"
}
res = requests.post(TOKEN_URL, data = data, headers = headers)
res_json = res.json()
TOKEN = res_json['result']['access_token']
```

```python
# 원본 템플릿 소유자 유저 아이디
PREV_USER_ID = ''

# 복제 받는 유저 아이디 리스트
NEXT_USER_ID = ['', '']

# 복제해줄 템플릿 아이디 리스트
TEMPLATE_IDS = ['', '', '']
```

```python
# 아래 코드는 수정하지 마세요.
# 아래 코드를 실행하려면, 코드를 클릭하고 [shift + enter]를 동시에 눌러주세요.
# 코드실행 결과는 코드 아래쪽에 결과가 나타납니다.
```

```python
URL = 'https://template.api.modusign.co.kr/v1/admin/copy/templates'
AUTHORIZATION = "Bearer " + TOKEN
final_result = []
failed_count = 0
process_count = 0

def alert() : 
    global process_count
    process_count += 1
    percent = process_count/len(NEXT_USER_ID)*100
    percent = round(percent, 2)
    print(str(percent) + "% 진행 중")
    

def template_copies(user_id) :
    result = []
    data = {
        "prevUserId": PREV_USER_ID,
        "nextUserId": user_id,
        "templateIds": TEMPLATE_IDS
    }
    data = json.dumps(data)
    headers = {
        "x-api-key": X_API_KEY,
        "contentType": "application/json",
        "authorization": AUTHORIZATION
    }
    res = requests.put(URL, data=data, headers=headers, timeout=5)
    alert()
    time.sleep(0.5)
    result.append(user_id)
    if res.status_code == 200:
        status_code_str = "성공"
        result.append(status_code_str)
    else :
        global failed_count
        failed_count += 1
        result.append(res.status_code)
    result.append(res.json())
    return final_result.append(result)


start_time = datetime.datetime.now()

for i in NEXT_USER_ID:
    template_copies(i)
    
finish_time = datetime.datetime.now()
finish_time_value = '{:%Y-%m-%d %H%M%S}'.format(finish_time)
print("========================================\n - 시작 시각은 " + start_time.strftime("%Y-%m-%d %H:%M:%S") + "이며, \n - 종료된 시각은 " + finish_time.strftime("%Y-%m-%d %H:%M:%S") + "입니다.\n")
print(' - 템플릿 복제가 실패된 유저수는 ' + str(failed_count) + ' 명 입니다.')
columns = ["유저", "성공여부", "응답결과"]
df = pd.DataFrame(final_result, columns = columns)
df.to_excel('result({0}).xlsx'.format(finish_time_value),sheet_name='result')
df
```
